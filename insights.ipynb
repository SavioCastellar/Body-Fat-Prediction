### Line Plot Feature Selection X Dataset

```python
def printLinePlot(feat_selection, data_partition, name):
    df_dp= df[df["data_partition"] == data_partition]
    df_dp_fs = df_dp[df_dp["feat_selection"] == feat_selection]
    fig = sns.lineplot(x='feat_set', y='MAPE', data=df_dp_fs, hue="modelo")
    plt.suptitle ("{} X Datasets".format(feat_selection), fontsize=12)
    plt.savefig("./imgs/{}.png".format(name))
```

```python
printLinePlot("Pearson", "<=100", "lineplot_pearson_less100")
```

```python
printLinePlot("Lasso", "<=100", "lineplot_lasso_less100")
```
